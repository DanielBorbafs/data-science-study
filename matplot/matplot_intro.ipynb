## plot()
essa √© uma fun√ß√£o que usamos para desenhar pontos(marcadores) em um diagrama.
por padr√£o ela dsenha uma linha de ponto a ponto.<br>
``plt.plot(parametro_1, parametro_2)``<br>

O Par√¢metro 1 √© uma matriz contendo os pontos do eixo x <br>
O Par√¢metro 2 √© uma matriz contendo os pontos do eixo y <br>

Exemplo:

Desenhe uma linha em um diagrama da posi√ß√£o (1, 3) para a posi√ß√£o (8, 10):


```python
eixo_x = [1,8]
eixo_y = [3,10]

plt.plot(eixo_x, eixo_y, color='#9EBC8A')
plt.show()
```

```python
# imagine que queremos usar pontos 'o' ao inv√©s de linhas, fazemos dessa forma:

plt.plot(eixo_x, eixo_y, 'o')
plt.show()
```

Podemos tamb√©m tra√ßar pontos m√∫ltiplos

Desenhe uma linha em um diagrama da posi√ß√£o (1, 3) para (2, 8), depois para (6, 1) e finalmente para a posi√ß√£o (8, 10):


```python
xpoints = [1,2,6,8]
ypoints = [3,8,1,10]

# Marcando os picos, ms = altera o tamanho, mec= bordas
# mfc= cores de dentro da borda
plt.plot(xpoints, ypoints, marker='o', ms=10, mec='b', mfc='green')
plt.show()
```

```python
plt.plot(xpoints, ypoints, 'o:g') # Marcando os picos utilizando fmt => marker,line,color

plt.show()
```

## Estilo de linha
podemos usar o argumento ``linestyle ou ls`` para alterar o estilo da linha plotada.


```python
ypoints = [1,5,9,12,16,4]

# tipos de linhas: dashed, dashdot, solid
plt.plot(ypoints, linestyle = 'dotted', marker='o', color='#9EBC8A') # definindo a cor
plt.show()
```

Para alterar a largura da linha usamos ``linewidth ou lw`` o valor vai ser um n√∫mero flutuante

```python
pontos_y = [3,8,1,10]

plt.plot(pontos_y, linewidth='12.5', color='#73946B')
plt.show()
```

<strong> Linhas m√∫ltiplas </strong>

Podemos adicionar quantas linhas quisermos apenasa adicionando mais ``plt.plot()``<br>

***imagine que precisamos tra√ßar duas linhas comparando gasto vs lucros***

```python
import matplotlib.pyplot as plt

gastos = [44500,48000 , 45555, 25600]
lucro = [24500, 35600, 65125, 81240]
meses = ['jan', 'fev', 'mar', 'may']

plt.plot(meses, gastos , color='#73946B', linestyle='dotted')
plt.plot(meses, lucro, color='g', linestyle='solid')
plt.show()
```

Devemos sempre transmitir clareza em nossos gr√°ficos, por isso √© importante sempre est√° colocando t√≠tulos, legendas e etc..<br>

Com matplot √© poss√≠vel definir tamb√©m a propriedade de fonte para t√≠tulo e r√≥tulos


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

# definindo as fontes
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.plot(x, y, color='#9EBC8A')

# t√≠tulos e r√≥tulos
plt.title('Gr√°fico de calorias queimas por pulso', loc='left') # Loc: define posi√ß√£o do texto
plt.xlabel("Pulso")
plt.ylabel("Calorias queimadas")

plt.show()
```


### linhas de grade
Com o Pyplot, voc√™ pode usar a ``grid()`` para adicionar linhas de grade ao gr√°fico.

```python
import matplotlib.pyplot as plt

x = [80, 85, 90, 95, 100, 105, 110, 115, 120, 125]
y = [240, 250, 260, 270, 280, 290, 300, 310, 320, 330]

plt.title('Dados de esportes')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x,y, color='#9EBC8A')

plt.grid(axis='x', color='#9EBC8A', linestyle = '--', linewidth = 0.5)
# ou axis = 'x' ou 'y' para exibir linha de grade das mesmas
# Tambem √© possivel definir linestyle, linewidth = 0.5, e cor

plt.show()

```

### subplot()

com essa fun√ß√£o conseguimos desenhar v√°rios gr√°ficos em uma figura.

```python
import matplotlib.pyplot as plt

# plot 01
mes = ['jan', 'fev', 'mar']
salario = [2200, 2500, 2800]
plt.subplot(1,2,1)
plt.plot(mes,salario, color='#9EBC8A')
plt.title('2024')

# plot 02
mes = ['jan', 'fev', 'mar']
salario = [3200, 4500, 6300]
plt.subplot(1,2,2)
plt.plot(mes,salario, color='#9EBC8A')
plt.title('2025')


plt.suptitle('Lucro primeiro trimestre')
plt.show()
```

``subplot()`` recebe tr√™s argumentos para definir o layout, que √© organizado em linhas em coluna, que s√£o representados pelo primeiro argumento


### Dispers√£o

com o pyplot usamos ``scatter()`` para desenhar um gr√°fico de dispers√£o.
Ela precisa de duas matrizes de mesmo comprimento, uma para os valores do eixo x e outra do eixo y

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y, color='#9EBC8A')
plt.show()
```

Nesse gr√°fico acima estamos comparando 13 carros, onde o eixo x √© a idade e o Y √© a velocidade, podemos observar que quanto mais novo o carro mais r√°pido ele corre.<br>

Tamb√©m conseguimos plotar dois gr√°ficos ao mesmo tempo nessa disper√ß√£o, vamos supor que agora temos dados do dia 1 e do dia 2.

```python
import matplotlib.pyplot as plt
import numpy as np

#Dia 01
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color='#537D5D', label='Dia 01')


#Dia02
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color='#9EBC8A', label='Dia 02')

plt.title('Velocidade de carros comparado a sua idade', pad=10)
plt.legend()
plt.show()
```

### Barras

### Histogramas

### Gr√°ficos de pizza
Com o pyplot, podemos usar `pie()` para desenhar gr√°ficos de pizza:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,25,25,15])
labels = ['Ma√ß√£', 'Banana', 'Amora', 'Abacaxi']
cores = ['#537D5D', '#73946B', '#9EBC8A', '#D2D0A0']
explod=[0.05, 0.0,0.0 ,0.0]

plt.figure(figsize=(10,6))
plt.pie(y, labels=labels, colors=cores, startangle=90, shadow=True, explode=explod)
plt.legend()

plt.show()

```
