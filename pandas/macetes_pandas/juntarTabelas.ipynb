merge() - Juntar tabelas com uma chave comum

Se as tabelas possuem uma coluna em comum, use essa coluna como chave:

```python
import pandas as pd 

df_clientes = pd.DataFrame({
    'ID': [1, 2, 3],
    'Nome': ['Alice', 'Bob', 'Charlie']
})

df_compras = pd.DataFrame({
      'ID': [1, 2, 3],
    'Compras': [200, 150, 300]
})

df_vendas = pd.merge(df_clientes, df_compras, on='ID') # INNER JOIN PADR√ÉO

print(df_vendas)
```

Juntando planilhas diferentes

```python
import pandas as pd 

df_produtos = pd.read_csv('produtos.csv')
df_vendas = pd.read_csv('vendas.csv')

df_vendasTotais = pd.merge(df_produtos, df_vendas, on='ProductID', how='left')

df_vendasTotais['TotalSalue'] = df_vendasTotais['Quantity'] * df_vendasTotais['SaleValue']

df_vendasTotais
```
