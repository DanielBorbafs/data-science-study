```python
import matplotlib.pyplot as plt

clientesCadastrados = [1100, 1115, 1250, 1332, 1844, 1962, 2150 ]
meses = ['jan', 'fev', 'mar', 'mai', 'jun', 'jul', 'ago']

plt.plot(meses, clientesCadastrados, marker='o', color='skyblue',)
plt.ylabel('quantidade')
plt.title('Evolução de clientes cadastrados - 2024')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dados
profissao = ['dev web', 'ux designer', 'dev backend']
salario = [5500, 6100, 4800]

# Criando DataFrame
df = pd.DataFrame(list(zip(profissao, salario)), columns=['profissao', 'salario'])

# Plotando o gráfico de barras
ax = df.plot('profissao', 'salario', kind='bar')

# Título do gráfico
plt.title('Salario por Profissao')

# Remover título do eixo X
plt.xlabel('')

# Remover legenda
plt.legend().set_visible(False)

# Ajustar rotação dos rótulos do eixo X
plt.xticks(rotation=0)

# Exibir o gráfico
plt.show()
```
